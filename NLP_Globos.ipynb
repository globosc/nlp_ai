1. **Anlisis de la Semntica Latente | LSA:**
Es una tcnica que utiliza procesos estadsticos para extraer y representar los significados de las palabras. Los significados son representados en trminos de su semejanza con otras palabras en un extenso texto llamado "corpus de documentos".

El LSA comienza procesando un texto de grandes dimensiones. Este texto contiene miles de prrafos y frases y es conocido como corpus lingstico. El corpus se representa en una matriz cuyas filas contiene todos los trminos distintos del corpus (palabras). De este modo, la matriz contiene sencillamente el nmero de veces que cada trmino aparece en un documento. Sobre esta matriz de frecuencias se efecta una ponderacin con el objeto de restar importancia a las palabras excesivamente frecuentes y aumentarla a las palabras moderadamente infrecuentes. La razn de esta ponderacin es sencilla: las palabras demasiado frecuentes no sirven para discriminar bien la informacin importante del prrafo y las moderadamente infrecuentes s. El siguiente paso es someter 16 esta matriz ponderada a un algoritmo llamado Descomposicin en Valores Singulares. SVD es una tcnica de reduccin de dimensiones.

1. **LexRank:**
Es un sistema de resmenes automtico no supervisado basado en grafos y el cual est inspiradoen tcnicas de PageRank, o sea: Ranking Lxico. 
Se basa en la idea de que la importancia de una frase depende de la similitud que esta tenga con el resto de oraciones del texto. 
Se basa en los clculos del coseno para calcular la similitud entre dos frases. Dos frases estn conectadas si el coseno de la similitud entre ambas es superior a cierto umbral. Para la seleccin de las frases que forman parte del resumen, busca cules son los nodos oraciones ms centrales del texto; es por ello que existen diversas medidas de centralidad las cuales pasaremos a comentar:
Centralidad basada en el grado: Donde la importancia de una frase depende del nmero de conexiones que esta tenga con el resto de nodos. 
Centralidad basada en vectores propios: Pondera cada una de las conexiones con la importancia del nodo con el que se conecta; por ello un nodo ser ms importante cuanto mayor sea el valor de los nodos a los que se conecta.

```python
import sumy
from sumy.parsers.plaintext import PlaintextParser
from sumy.nlp.tokenizers import Tokenizer
from sumy.summarizers.lex_rank import LexRankSummarizer
from sumy.summarizers.lsa import LsaSummarizer
from newspaper import Article
```

### Contexto

Para las siguientes partes de cdigo, vamos a utilizar un poco de Web Scraping para extraer el contenido de la ley publicada en el diario oficial del Congreso Colombiano.

Para este ejemplo se utiliza el Congreso de este pas dado que su publicacin es a travs de una pgina scrapeable ya que es html puro, no as el caso de Chile en donde la Biblioteca Nacional y el Diario Oficial enmascaran con Javascript. 

```python
#Caso de Estudio 1 
ley1 = "https://www.suin-juriscol.gov.co/viewDocument.asp?id=30039047"
```

```python
# Descargar el articulo y convertirlo a texto plano en un gran string
contenido1 = Article(ley1)
contenido1.download()
contenido1.parse()
 
# articulo
print(contenido1.text)
```

```python
#Vamos a tokenizar nuestro texto utilizando nuestro lenguaje natural
parser = PlaintextParser.from_string(contenido1.text,Tokenizer("spanish"))
parser
```

```python
# Usando LexRank
lexrank = LexRankSummarizer()
# Usando LSA
lsa = LsaSummarizer()
#Resumir el proyecto de ley en 3 ideas
resumen_lexrank = lexrank(parser.document, 3)
resumen_lsa = lsa(parser.document, 3)
```

```python
start = "\033[1m"
end = "\033[0;0m"
```

```python
print(start + 'Resumen utilizando algoritmo LexRank:' + end)
print()
for ideas in resumen_lexrank:
    print(ideas, '\n')
```

```python
print(start + 'Resumen utilizando algoritmo LSA:' + end)
print()
for ideas in resumen_lsa:
    print(ideas, '\n')
```
